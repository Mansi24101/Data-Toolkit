2. How does broadcasting work in NumPy?
- Broadcasting is a powerful mechanism in NumPy that allows arrays of different shapes to be used in arithmetic operations together, without explicit replication of data.
It automatically expands smaller arrays to match the shape of larger ones during element-wise operations.
- To apply broadcasting, NumPy compares shapes from right to left. Two dimensions are compatible if:
They are equal, or
One of them is 1
- Broadcasting reduces memory use.
Avoids explicit loops.
Enables fast vectorized operations.

4. Explain the use of the groupby() method in Pandas?
- The groupby() method in Pandas is used to split a DataFrame into groups based on some criteria, then apply a function (like mean, sum, count, etc.), and combine the results. This is often referred to as the "split-apply-combine" strategy.
-  It’s extremely useful for:
Aggregating data,
Analyzing subgroups,
Summarizing patterns across categories.
-  Common Aggregation Functions with groupby():  .mean() – average,  .sum() – total,   .count() – number of entries,  .max() / .min() – highest/lowest,  .agg() – apply multiple functions


```python
import pandas as pd
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'],
    'Employee': ['Mansi', 'Neha', 'Abhi', 'Rashu', 'Shaun'],
    'Salary': [50000, 55000, 60000, 65000, 70000]
}

df = pd.DataFrame(data)

avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

6. What are the differences between NumPy arrays and Python lists?
- The differences between NumPy arrays and Python lists:
- 1. Data Type Consistency:

NumPy Array: Homogeneous — all elements must be of the same type.

Python List: Heterogeneous — can contain elements of different data types.
Syntax/example: np_array = np.array([1, 2, 3])     # All integers

py_list = [1, 'a', 3.5]            # Mixed types

- 2. Performance:

NumPy Array: Faster due to efficient memory storage and vectorized operations written in C.

Python List: Slower for large numerical operations.

- 3. Functionality:

NumPy Array: Supports advanced mathematical operations (e.g., broadcasting, linear algebra, FFT).

Python List: Limited to basic operations like append, remove, sort.

- 4. Memory Usage:

NumPy Array: More memory efficient due to compact storage.

Python List: Uses more memory due to storing references and type metadata.

-  5. Vectorization:

NumPy Array: Can perform operations on the whole array without loops.
example/syntax: np_array * 2  # Multiplies each element

Python List: Requires explicit loops or list comprehensions.
example/syntax: [x * 2 for x in py_list]

```python
import numpy as np
np_array = np.array([1, 2, 3, 4, 5])

np_result = np_array * 5

print("NumPy array result:", np_result)
```

```python
py_list = [1, 2, 3, 4, 5]
py_result = [x * 5 for x in py_list]

print("Python list result:", py_result)
```

7. What is a heatmap, and when should it be used?
- A heatmap is a data visualization technique that uses color to represent the magnitude of values in a matrix or 2D array. It makes it easier to spot patterns, correlations, or anomalies in large datasets.

- Heatmap should be used when we want to:

Visualize correlation between features (e.g., in a correlation matrix).

Spot trends or patterns in a dataset (e.g., time-series or spatial data).

Compare values across two dimensions (e.g., products vs. months, genes vs. samples).

Identify clusters or outliers in data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Math': [80, 90, 70, 65],
    'Science': [85, 95, 75, 60],
    'English': [78, 88, 72, 68]
}
df = pd.DataFrame(data)

corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Subject Score Correlation Heatmap")
plt.show()
```

21. Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy’s array slicing is significantly faster than slicing Python lists because of how NumPy arrays are implemented under the hood. Here’s why:

- Key Reasons Why NumPy Slicing is Faster:
1. Contiguous Memory Allocation
NumPy arrays are stored in a contiguous block of memory, unlike Python lists, which are arrays of pointers to separate memory locations.
This allows NumPy to access and slice data using low-level memory operations that are much faster.

2. No Type Checking Overhead
Python lists can store elements of different types, so each access requires type checking.
NumPy arrays are homogeneous (same data type), eliminating the need for repeated type checking during slicing.

3. Views Instead of Copies
NumPy slicing returns a view (a reference) to the original data whenever possible, not a copy.
Python list slicing creates a new list every time, which takes more time and memory.
4. C-Level Optimizations
NumPy is implemented in C, and its slicing operations are heavily optimized using compiled code.
Python lists use slower interpreted operations in Python's runtime.

5. Vectorized Operations
NumPy supports vectorized operations, allowing bulk operations without explicit loops.
Python lists require loops for element-wise operations, which are slower.

```python
import numpy as np
import time
arr = np.arange(1_000_000)
start = time.time()
_ = arr[1000:900000]
print("NumPy slicing time:", time.time() - start)

lst = list(range(1_000_000))
start = time.time()
_ = lst[1000:900000]
print("List slicing time:", time.time() - start)
```

1. How do you create a 2D NumPy array and calculate the sum of each row?

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'height': [150, 160, 170, 180, 190],
    'weight': [50, 60, 70, 80, 90],
    'age': [25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

8. Create a line plot with multiple lines using Matplotlib?

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]
y3 = [25, 16, 9, 4, 1]

plt.plot(x, y1, label='y = x²', color='blue', marker='o')
plt.plot(x, y2, label='y = x', color='green', linestyle='--')
plt.plot(x, y3, label='y = reversed x²', color='red', linestyle='-.')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Plot Example')
plt.legend()
plt.grid(True)

plt.show()
```

13. Create a 3D scatter plot using Plotly?


```python
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [100, 101, 102, 103, 104]

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(255, 0, 0)',
        opacity=0.8
    )
)])

fig.update_layout(
    title="3D Scatter Plot",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

fig.show()
```
